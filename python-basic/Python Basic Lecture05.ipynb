## 문자열(String)
- 문자열(string)
    - 문자(character)들의 모임
- 문자(character)
    - 영문자: a ~ z, A ~ Z, 그 외 문자들
    - 특수문자: !@#$.,
    - 공백문자(whitespace character): tab, enter(newline, 줄바꿈문자), space(공백문자)
   
- 작은 따옴표 : `''`
- 큰 따옴표 : `""`

문자열으로 둘다 가능함

### 탈출문자 표현법
|Escape code| 설명|
|---|---|
|\n| 줄바꿈(=개행)|
|\t |수평 탭문자(8칸)|
|\\ |문자 "\"|
|\' |작은따옴표(')|
|\" |큰따옴표(")|
|\r |캐리지 리턴|
|\f |폼 피드|
|\a |벨 소리|
|\b |백 스페이스|
|\000 |널문자|

### Raw 문자열
- 탈출문자를 모두 그대로의 문자로 취급
- 문자열 앞에 r 붙히기

````python
ec=r"Life is too short\nYou need python\n"
print(ec)
````

````python
s.index('too') # too의 index
````

````python
comma = ','
comma.join(s)
````

````python
# 실질적인 예시는 다음과 같다
x = ['서울시','서대문구','신촌동']
print(' '.join(x))
````

### 9. format 함수

````python
"풍속 {0}ms/s".format(5)
````

````python
"풍속 {0}ms/s".format("five")
````

````python
wind = 9
"풍속 {}ms/s".format(wind)
````

````python
"{1} {0}ms/s".format( 5, '속') # index로 위치 지정 가능
````

````python
"{title} {wind}ms/s".format( wind= 15 , title='풍속')
````

````python
"{title} {0}ms/s".format(5, title='풍속!!')
````

````python
"{0:10}".format("hello")  # 여백
````

````python
 "{0:<10}".format("hello")
````

````python
"{0:>10}".format("hello")
````

````python
"{0:^10}".format("hello")
````
