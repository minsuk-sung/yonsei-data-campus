#### 8)  리눅스 명령어를 사용하여 보자
- 실제 terminal에서 pipe를 이용한 결과 분석도 자주 이용됨
- pipe (|) 를 통하여 결과를 다음 명령어에 전달 가능 (windows도)
- 알아두면 좋은 linux 필수 명령어
 + awk - 주어진 형태로 주어진 line을 파싱함
 + cat - 파일을 출력
 + cut - 주어진 출력을 특정 기준으로 분리함
 + grep - 주어진 출력에서 특정 패턴을 가지고 있는지 확인
 + head - 파일의 시작부분을 출력함
 + sed - 주어진 패턴을 지정한 패턴으로 치환함
 + tail - 파일의 끝부분을 출력함
 + touch - 빈 파일을 생성함, 존재하는 파일의 수정 시간을 변경
 + uniq - 출력 중 unique 한 결과만 출력함 (sort 후 사용)
 + wc - 출력의 개수를 집계할 때 쓰임

```python
!cut --help
```

```python
#윈도우 폴더에 있는 파일들을 ls를 이용하여 출력하기
!ls -al c:/windows
```

```python
#특정 파일을 cat으로 출력하기
!cat c:/windows/setupact.log
```

```python
#head를 이용하여 특정파일 내용을 상위 10개 출력
!head c:/windows/setupact.log -n10
```

```python
#tail을 이용하여 특정파일 내용을 하위 10개 출력
!tail c:/windows/setupact.log -n10
```

```python
#grep을 이용하여 특정 패턴이 포함된 line을 출력
!grep 00010001  c:/windows/setupact.log
```

```python
#sed를 이용하여 특정 패턴을 주어진 패턴으로 치환
!sed s/00010001/bigdataworld/g c:/windows/setupact.log
```

```python
#awk을 이용하여 공백으로 분할된 column 중 첫번째를 출력
!awk "{print $1}" c:/windows/setupact.log
```

```python
#wc를 이용하여 특정 파일의 line수, 글자수, 단어수를 출력
!wc c:/windows/setupact.log
```
