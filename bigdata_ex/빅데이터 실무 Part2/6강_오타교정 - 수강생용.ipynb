## 7. Qwerty 오타 치환

* 영한 전환 오타를 일반화
 * 한글은 영타 자판과 1:1 매핑하지 않음
 * 그러나 자소는 영타 자판과 1:1 매핑함
  * jaso(한글) = ㅎㅏㄴㄱㅡㄹ
  * Qwerty(ㅎㅏㄴㄱㅡㄹ) = gksrmf
 ---------------------------------
* qwerty 오타 공식
 * 정타 != 오타 and qwerty(jaso(정타)) == qwerty(jaso(오타))
 * True이면 qwerty 전환 오타

```python
#해당 함수는 이전대비 된소리에 대한 처리가 포함되어있음
#모든 한글에 대한 처리를 하기 위함
def decomposeKStr(kstr):
    def decomposeKChar(ch):
        if not(0xAC00 <= ord(ch) <= 0xD7AF):
            return ch

        cho  = ["ㄱ", "ㄲ", "ㄴ", "ㄷ", "ㄸ", "ㄹ", "ㅁ", "ㅂ", "ㅃ", "ㅅ", "ㅆ", "ㅇ", "ㅈ", "ㅉ", "ㅊ", "ㅋ", "ㅌ", "ㅍ", "ㅎ" ]
        jung = ["ㅏ", "ㅐ", "ㅑ", "ㅒ", "ㅓ", "ㅔ", "ㅕ", "ㅖ", "ㅗ", "ㅘ", "ㅙ", "ㅚ", "ㅛ", "ㅜ", "ㅝ", "ㅞ", "ㅟ", "ㅠ", "ㅡ", "ㅢ", "ㅣ"]
        jung_double = {"ㅘ":"ㅗㅏ", "ㅙ":"ㅗㅐ", "ㅚ":"ㅗㅣ", "ㅝ":"ㅜㅓ", "ㅞ":"ㅜㅔ", "ㅟ":"ㅜㅣ", "ㅢ":"ㅡㅣ"}
        jong = ["", "ㄱ", "ㄲ", "ㄳ", "ㄴ", "ㄵ", "ㄶ", "ㄷ", "ㄹ", "ㄺ", "ㄻ", "ㄼ", "ㄽ", "ㄾ", "ㄿ", "ㅀ", "ㅁ", "ㅂ", "ㅄ", "ㅅ", "ㅆ", "ㅇ", "ㅈ", "ㅊ", "ㅋ", "ㅌ", "ㅍ", "ㅎ"]
        jong_double = {"ㄳ":"ㄱㅅ", "ㄵ":"ㄴㅈ", "ㄶ":"ㄴㅎ", "ㄺ":"ㄹㄱ", "ㄻ":"ㄹㅁ", "ㄼ":"ㄹㅂ", "ㄽ":"ㄹㅅ",
                       "ㄾ":"ㄹㅌ", "ㄿ":"ㄹㅍ", "ㅀ":"ㄹㅎ", "ㅄ":"ㅂㅅ"
                      }

        n = ord(ch) - 0xAC00

        last = n % 28
        middle = ((n-last) // 28) % 21
        first = (((n-last) // 28) - middle) // 21
        return cho[first], jung_double.get(jung[middle], jung[middle]), jong_double.get(jong[last], jong[last])
    
    res = []
    for ch in kstr:
        res += decomposeKChar(ch)
    return "".join(res)
```

```python
#jaso -> qwerty 전환 함수
def qweryKStr(kstr):
    qwerty = "qwertyuiopasdfghjklzxcvbnmQWERTOP"
    jaso = "ㅂㅈㄷㄱㅅㅛㅕㅑㅐㅔㅁㄴㅇㄹㅎㅗㅓㅏㅣㅋㅌㅊㅍㅠㅜㅡㅃㅉㄸㄲㅆㅒㅖ"
    
    table = {j:q for j,q in zip(jaso, qwerty)}
    return "".join(table.get(ch, ch) for ch in decomposeKStr(kstr))
```

```python
print(qweryKStr("네이버"))
print(qweryKStr("spdlqj"))
```

```python
#edit distance 함수
import numpy as np
def levenshtein(seq1, seq2):  
    size_x = len(seq1) + 1
    size_y = len(seq2) + 1
    matrix = np.zeros ((size_x, size_y))
    for x in range(size_x):
        matrix [x, 0] = x
    for y in range(size_y):
        matrix [0, y] = y

    for x in range(1, size_x):
        for y in range(1, size_y):
            if seq1[x-1] == seq2[y-1]:
                matrix [x,y] = min(
                    matrix[x-1, y] + 1,
                    matrix[x-1, y-1],
                    matrix[x, y-1] + 1
                )
            else:
                matrix [x,y] = min(
                    matrix[x-1,y] + 1,
                    matrix[x-1,y-1] + 1,
                    matrix[x,y-1] + 1
                )
    return int(matrix[size_x - 1, size_y - 1])
```

```python
print(levenshtein("네이버", "네이버"))
print(levenshtein("넹이버", "네이버"))
print(levenshtein("다음", "네이버"))
```

## 8. pandas를 활용한 오타교정 (사전형)
- 연관검색어 로직을 이용하여 느슨한 연관 검색어 집합을 생성

```python
#이전 사용했던 연관 검색어를 추출하는 로직을 그대로 사용함
#df7까지 수행하면 A, B, key_A, key_B를 가지는 df가 나옴
import pandas as pd
import csv 
df = pd.read_csv('data/sample_search_log.txt', sep='\t', header=None, names = ['date', 'id', 'keyword', 'result'],
                 quoting=csv.QUOTE_NONE)
df =  df[df["date"].map(str).map(len) == 15]
df['date'] = df.date.apply(lambda x: pd.datetime.strptime(x, '%Y%m%d-%H%M%S'))
df = df[['date', 'id', 'keyword']]
df = df.dropna()
df = df.sort_values(by=['date', 'id'])
gdf = df.groupby('id').filter(lambda x:x['keyword'].count() > 2).groupby('id')
df2 = gdf.agg({'keyword':list})
df3 = df2.keyword.apply(lambda x:[tuple(x[i:i+2]) for i in range(len(x)-1)]).to_frame()
df4 = df3.keyword.apply(pd.Series)
df5 = df4.stack().to_frame('pairwise')
df6 = df5.reset_index(level=0)
df7 = df6.pairwise.apply(lambda x:x[0]).to_frame('A')
df7['B'] = df6.pairwise.apply(lambda x:x[1])
df7['key_A'] = df7.A.apply(lambda x:x.replace(' ','').lower())
df7['key_B'] = df7.B.apply(lambda x:x.replace(' ','').lower())
df7.head(10)
```

```python
#qwerty를 수행시켜 각 key_A, key_B에 대해서 qwerty화 시킨다
df7_copy = df7.copy()
df7_copy['qwert_A'] = df7_copy.key_A.map(qweryKStr)
df7_copy['qwert_B'] = df7_copy.key_B.map(qweryKStr)
df7_copy.head(10)
```

```python
# qwerty_A와 qwerty_B에 edit distance 동작을 수행한다
df8 = df7_copy[(df7_copy.qwert_A.str.len() < 20) & (df7_copy.qwert_B.str.len() < 20)]
df8['qwert_ed'] = df8[['qwert_A', 'qwert_B']].apply(lambda x:levenshtein(*x), axis=1)
df8.head(10)
```

```python
# 러프한 오타 후보를 포함한 df 를 추출해본다
# ed < 충분히 큰 거리 & 오타와 정타 키워드가 같지 않은 필드
df9 = df8[(df8.qwert_ed < 6) & (df8.key_A != df8.key_B)]
df9.head(10)
```

```python
# key_A 쪽 확률 P(key_A)를 구하기
gdf = df9.groupby(by='key_A')[['key_A']]
probA_df9 = gdf.count().rename(columns={'key_A':'cnt_A'}).sort_values(by='cnt_A', ascending=False)
probA_df9['prob_A'] = probA_df9.cnt_A/probA_df9.cnt_A.sum()
probA_df9.head(5)
```

```python
# key_B 쪽 확률 P(key_B)를 구하기
gdf = df9.groupby(by='key_B')[['key_B']]
probB_df9 = gdf.count().rename(columns={'key_B':'cnt_B'}).sort_values(by='cnt_B', ascending=False)
probB_df9['prob_B'] = probB_df9.cnt_B/probB_df9.cnt_B.sum()
probB_df9.head(5)
```

```python
# 각 확률 P(key_A) 와 P(key_B)를 조인한다
df10 = df9.join(probA_df9.prob_A, on='key_A', how='left').join(probB_df9.prob_B, on='key_B', how='left')
df10.head(5)
```

```python
# 오타 일경우 P(key_A) < P(key_B)가 성립한다고 가정
# 수행한다
df11 = df10[df10.prob_A < df10.prob_B]
df11.head(10)
```

```python
# edit distance를 엄격하게 수행해본다 1 or 2
df12 = df11[df11.qwert_ed < 2]
df12.head(10)
```

```python
# 명백한 오타를 구할 수 있음
# qwerty 는 서로 같지만 입력은 다른 경우
typo_type1 = df12[(df12.key_A != df12.key_B) & (df12.qwert_A == df12.qwert_B)].sort_values(by='prob_B', ascending=False)
typo_type1['type'] = '부분오타형'
typo_type1.head(10)
```

```python
# 명백한 오타를 구할 수 있음
# 사전 입력과 사전 qwerty 는 서로 같지만 사후 입력과 사후 qwert는 다른 경우
typo_type2 = df12[(df12.key_A == df12.qwert_A) & (df12.key_B != df12.qwert_B)].sort_values(by='prob_B', ascending=False)
typo_type2['type'] = '치환형'
typo_type2.head(10)
```

```python
# 사전 key 도 qwerty와 다르고 사후 key 도 qwerty와 다르면 명박하게 어려운 오타가 된다 
type_type3 = df12[(df12.key_A != df12.qwert_A) & (df12.key_B != df12.qwert_B)].sort_values(by='prob_B', ascending=False)
type_type3['type'] = '확률형'
type_type3.head(10)
```

```python
6# 오타의 유형을 union해서 통합된 결과로 만든다
res = pd.concat([typo_type1[['A', 'B', 'type']], typo_type2[['A', 'B', 'type']], type_type3[['A', 'B', 'type']] ])
res = res.set_index('A')
res.head(10)
```

```python
# 테스트를 하기 위해서 메모리에 저장
res2= {key:val for key, val in zip(res.index, [(x, y) for x, y in zip(res.B, res.type)])}
```
