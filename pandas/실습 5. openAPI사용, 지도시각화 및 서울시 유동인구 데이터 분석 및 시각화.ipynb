```python
map1 = folium.Map(location = [37.566535, 126.97796919999996], zoom_start = 15)
map1
# 서울시청을 지도의 중심으로, 지도를 그려줌. 
# zoom_start는 지도를 그릴 때 얼마나 zoom 해서 보여줄 지를 결정.  값이 클수록, zoom in
```

```python
# Marker 추가하기 (지도에 지점 추가하기)
folium.Marker([37.566535, 126.97796919999996],popup='테스트 지점').add_to(map1)
map1
```

```python
josa = pd.read_excel('./data/서울시유동인구/4_유동인구_조사지점정보_2015.xlsx',
                       skiprows = [0,1,3])
josa
```

```python
len(josa)
```

```python
# 필요한 컬럼만 선택
# 30개만 선택해서 지도에 마커로 추가
josa30 = josa[['조사지점명','X좌표','Y좌표']].iloc[0:30]
josa30
```

```python
# 위치 정보를 인코딩하는 방식이 매우 다양
# X좌표, Y좌표의 인코딩 방식을 변경할 필요가 있음
# 이를 위한 라이브러리인 pyproj를 설치하고 import
# !pip install pyproj
from pyproj import Proj, transform
```

```python
# 위치 정보를 인코딩하는 변환 함수를 정의
def convert_loc(x,y):
    inProj = Proj(init = 'epsg:5181') # 인코딩 방식
    outProj = Proj(init = 'epsg:4326')
    x2, y2 = transform(inProj,outProj,x,y)
    return [y2, x2]
```

```python
josa30['위경도'] = josa30.apply(lambda row: convert_loc(row.X좌표, row.Y좌표),axis = 1)
josa30
```

```python
# 30개의 조사지점을 각각 마커로 생성하여 맵에 추가
for index, row in josa30.iterrows():
    print(row.조사지점명, row.위경도)
    folium.Marker(row.위경도, popup = row.조사지점명).add_to(map1)
```
